```python

import binascii
from Crypto.Util.Padding import pad

# Group Members: menna kurdi 13005718  ,Jana saeed 13003945 ,aya moustafa 13001223



# DES Tables
IP = [58, 50, 42, 34, 26, 18, 10, 2, 60, 52, 44, 36, 28, 20, 12, 4,
      62, 54, 46, 38, 30, 22, 14, 6, 64, 56, 48, 40, 32, 24, 16, 8,
      57, 49, 41, 33, 25, 17, 9, 1, 59, 51, 43, 35, 27, 19, 11, 3,
      61, 53, 45, 37, 29, 21, 13, 5, 63, 55, 47, 39, 31, 23, 15, 7]

FP = [40, 8, 48, 16, 56, 24, 64, 32, 39, 7, 47, 15, 55, 23, 63, 31,
      38, 6, 46, 14, 54, 22, 62, 30, 37, 5, 45, 13, 53, 21, 61, 29,
      36, 4, 44, 12, 52, 20, 60, 28, 35, 3, 43, 11, 51, 19, 59, 27,
      34, 2, 42, 10, 50, 18, 58, 26, 33, 1, 41, 9, 49, 17, 57, 25]

E = [32, 1, 2, 3, 4, 5, 4, 5, 6, 7, 8, 9, 8, 9, 10, 11, 12, 13, 12, 13, 14, 15, 16, 17,
     16, 17, 18, 19, 20, 21, 20, 21, 22, 23, 24, 25, 24, 25, 26, 27, 28, 29, 28, 29, 30, 31, 32, 1]

P = [16, 7, 20, 21, 29, 12, 28, 17, 1, 15, 23, 26, 5, 18, 31, 10,
     2, 8, 24, 14, 32, 27, 3, 9, 19, 13, 30, 6, 22, 11, 4, 25]

S_BOXES = [
    [[14, 4, 13, 1, 2, 15, 11, 8, 3, 10, 6, 12, 5, 9, 0, 7],
     [0, 15, 7, 4, 14, 2, 13, 1, 10, 6, 12, 11, 9, 5, 3, 8],
     [4, 1, 14, 8, 13, 6, 2, 11, 15, 12, 9, 7, 3, 10, 5, 0],
     [15, 12, 8, 2, 4, 9, 1, 7, 5, 11, 3, 14, 10, 0, 6, 13]],

    [[15, 1, 8, 14, 6, 11, 3, 4, 9, 7, 2, 13, 12, 0, 5, 10],
     [3, 13, 4, 7, 15, 2, 8, 14, 12, 0, 1, 10, 6, 9, 11, 5],
     [0, 14, 7, 11, 10, 4, 13, 1, 5, 8, 12, 6, 9, 3, 2, 15],
     [13, 8, 10, 1, 3, 15, 4, 2, 11, 6, 7, 12, 0, 5, 14, 9]],

    [[10, 0, 9, 14, 6, 3, 15, 5, 1, 13, 12, 7, 11, 4, 2, 8],
     [13, 7, 0, 9, 3, 4, 6, 10, 2, 8, 5, 14, 12, 11, 15, 1],
     [13, 6, 4, 9, 8, 15, 3, 0, 11, 1, 2, 12, 5, 10, 14, 7],
     [1, 10, 13, 0, 6, 9, 8, 7, 4, 15, 14, 3, 11, 5, 2, 12]],

    [[7, 13, 14, 3, 0, 6, 9, 10, 1, 2, 8, 5, 11, 12, 4, 15],
     [13, 8, 11, 5, 6, 15, 0, 3, 4, 7, 2, 12, 1, 10, 14, 9],
     [10, 6, 9, 0, 12, 11, 7, 13, 15, 1, 3, 14, 5, 2, 8, 4],
     [3, 15, 0, 6, 10, 1, 13, 8, 9, 4, 5, 11, 12, 7, 2, 14]],

    [[2, 12, 4, 1, 7, 10, 11, 6, 8, 5, 3, 15, 13, 0, 14, 9],
     [14, 11, 2, 12, 4, 7, 13, 1, 5, 0, 15, 10, 3, 9, 8, 6],
     [4, 2, 1, 11, 10, 13, 7, 8, 15, 9, 12, 5, 6, 3, 0, 14],
     [11, 8, 12, 7, 1, 14, 2, 13, 6, 15, 0, 9, 10, 4, 5, 3]],

    [[12, 1, 10, 15, 9, 2, 6, 8, 0, 13, 3, 4, 14, 7, 5, 11],
     [10, 15, 4, 2, 7, 12, 9, 5, 6, 1, 13, 14, 0, 11, 3, 8],
     [9, 14, 15, 5, 2, 8, 12, 3, 7, 0, 4, 10, 1, 13, 11, 6],
     [4, 3, 2, 12, 9, 5, 15, 10, 11, 14, 1, 7, 6, 0, 8, 13]],

    [[4, 11, 2, 14, 15, 0, 8, 13, 3, 12, 9, 7, 5, 10, 6, 1],
     [13, 0, 11, 7, 4, 9, 1, 10, 14, 3, 5, 12, 2, 15, 8, 6],
     [1, 4, 11, 13, 12, 3, 7, 14, 10, 15, 6, 8, 0, 5, 9, 2],
     [6, 11, 13, 8, 1, 4, 10, 7, 9, 5, 0, 15, 14, 2, 3, 12]],

    [[13, 2, 8, 4, 6, 15, 11, 1, 10, 9, 3, 14, 5, 0, 12, 7],
     [1, 15, 13, 8, 10, 3, 7, 4, 12, 5, 6, 11, 0, 14, 9, 2],
     [7, 11, 4, 1, 9, 12, 14, 2, 0, 6, 10, 13, 15, 3, 5, 8],
     [2, 1, 14, 7, 4, 10, 8, 13, 15, 12, 9, 0, 3, 5, 6, 11]]
]


# 16 round keys (48 bits each)
round_keys = ["010101010101010101010101010101010101010101010101" for _ in range(16)]

# Helper function for permutation
def permute(block, table):
    return ''.join(block[i - 1] for i in table)

# S-box substitution
def sbox_substitution(xor_result):
    sbox_output = ''
    for i in range(8):
        block = xor_result[i * 6:(i + 1) * 6]
        row = int(block[0] + block[5], 2)
        col = int(block[1:5], 2)
        val = S_BOXES[i][row][col]
        sbox_output += f"{val:04b}"
    return sbox_output

# DES encryption for a single 64-bit block
def des_encrypt_block(block, key):
    print(f"\nCBC XOR Block (hex): {block}")
    block = bin(int(block, 16))[2:].zfill(64)  # Convert to 64-bit binary
    print(f"Initial Permutation: {block}")
    block = permute(block, IP)  # Initial Permutation
    left, right = block[:32], block[32:]

    for round_number in range(16):
        print(f"\nRound {round_number + 1}:")
        right_expanded = permute(right, E)  # Expansion
        print(f" Expanded R:        {right_expanded}")
        round_key = round_keys[round_number]
        xor_result = bin(int(right_expanded, 2) ^ int(round_key, 2))[2:].zfill(48)  # Key Mixing
        print(f" XOR with key:      {xor_result}")

        sbox_output = sbox_substitution(xor_result)  # S-box substitution
        print(f" S-box output:      {sbox_output}")

        pbox_output = permute(sbox_output, P)  # P-box Permutation
        print(f" Permutation P:     {pbox_output}")

        xor_with_left = bin(int(left, 2) ^ int(pbox_output, 2))[2:].zfill(32)
        print(f" XOR with L:        {xor_with_left}")

        left, right = right, xor_with_left  # Swap halves
        print(f" New L:  {left}")
        print(f" New R:  {right}")
        print("-" * 48)

    combined = right + left  # Swap before final permutation
    print(f"\nBefore Final Permutation (swapped halves): {combined}")
    encrypted_block = permute(combined, FP)
    print(f"Final Permutation:                        {encrypted_block}")
    return hex(int(encrypted_block, 2))[2:].zfill(16)  # Ensure 16-character hex string

# CBC mode encryption
def cbc_encrypt(plaintext, key, iv):
    print("\nEncrypting plaintext using DES in CBC mode with detailed steps:")
    padded = pad(plaintext.encode(), 8)
    blocks = [padded[i:i+8] for i in range(0, len(padded), 8)]
    encrypted = []
    previous = iv

    for block in blocks:
        xor_block = bytes(a ^ b for a, b in zip(block, previous))  # XOR with IV or previous ciphertext
        encrypted_block = des_encrypt_block(binascii.hexlify(xor_block).decode(), key)
        encrypted.append(encrypted_block)
        previous = binascii.unhexlify(encrypted_block)  # Update previous block

    return ''.join(encrypted)

# Test
key = '133457799BBCDFF1'  # Example key
iv = 'a1b2c3d4e5f60708'  # Example IV
plaintext = "Hello DES in CBC Mode"

encrypted = cbc_encrypt(plaintext, key, binascii.unhexlify(iv))
print("\nFinal Results:")
print(f"Key:  {key}")
print(f"IV:   {iv}")
print(f"Ciphertext:  {encrypted}")
```
